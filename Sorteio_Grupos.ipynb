```python
import random
from IPython.core.display import display, HTML
import pandas as pd
import datetime
```

```python
grupos = {
  'Distribuição de Arquivos Multimídia':
  {'Integrantes':['A', 'B', 'C', 'D'],
   'Link':'https://meet.google.com/'},
  
  'Processamento de Multimídia': 	
  {'Integrantes':['A', 'B', 'C', 'D'],
   'Link':'meet.google.com/'},

  'Computação em Grade': 	
  {'Integrantes':['A', 'B', 'C', 'D'],
   'Link':'https://meet.google.com/'},

  'Blockchain': 	
  {'Integrantes':['A', 'B', 'C', 'D'],
   'Link':'meet.google.com/evs-vvwz-gby'},

  'Acesso Remoto': 	
  {'Integrantes':['A', 'B', 'C', 'D'],
   'Link':'https://meet.google.com/'},

  'Banco de Dados Distribuido': 
  {'Integrantes':['A', 'B', 'C', 'D'],
   'Link':'https://meet.google.com/'},
  
  'Busca' :
  {'Integrantes':['A', 'B', 'C', 'D'],
   'Link':'https://meet.google.com/ziu-upax-xof'},
  
  'Mensagem Criptografada':
  {'Integrantes':['A', 'B', 'C', 'D'],
   'Link':'meet.google.com/'}
}
```

```python
#%load_ext google.colab.data_table

import datetime
data = datetime.datetime.strptime("13/09/2021","%d/%m/%Y")
dia = datetime.timedelta(days=1)
semana = datetime.timedelta(days=7)
duracao_mentoria = datetime.timedelta(minutes=10)
horarios = [
            datetime.datetime.strptime("18:45","%H:%M"),
            datetime.datetime.strptime("20:35","%H:%M")
]
ultima_data = datetime.datetime.strptime("30/11/2021","%d/%m/%Y")
feriados = [
    datetime.datetime.strptime("11/10/2021","%d/%m/%Y"),
    datetime.datetime.strptime("12/10/2021","%d/%m/%Y"),
    datetime.datetime.strptime("1/11/2021","%d/%m/%Y"),
    datetime.datetime.strptime("2/11/2021","%d/%m/%Y"),
    datetime.datetime.strptime("15/11/2021","%d/%m/%Y"),

]
```

```python
def sort_n_display(data, horario, indice = [0]):
  if(indice[0] == 0):
    indice[0] += 1
    return ''
  html = ''
  
  if(data in feriados):
    html += '<h2>Dia %02d/%02d/%d</h2>'%(data.day,data.month,data.year)
    html += "<h1>Não havera aula - Feiado<h1>"
    display(HTML(html))
    return html
  ordem = list(grupos.keys())
  random.shuffle(ordem)

  final = {'Tema':[],'Integrantes':[],'Período':[]}
  inicio = horarios[horario]
  for i,tema in enumerate(ordem):
    #final['Nº da Fila'].append(i+1)
    ate = inicio+duracao_mentoria
    final['Período'].append("%02d:%02d~%02d:%02d"%(inicio.hour, inicio.minute, ate.hour,ate.minute))
    inicio+=duracao_mentoria
    final['Tema'].append('<a href="%s">%s</a>'%(grupos[tema]['Link'],tema))
    final['Integrantes'].append(','.join(grupos[tema]['Integrantes'])) 
  df = pd.DataFrame(final)
  
  html += '<h2>%dª Mentoria - dia %02d/%02d/%d</h2>'%(indice[0],data.day,data.month,data.year)
  indice[0] += 1
  html += df.to_html().replace('&lt;','<').replace('&gt;','>')
  display(HTML(html))
  return html

html = ''
while data <= ultima_data:
  html += sort_n_display(data,0)
  html += sort_n_display(data+dia,1)
  data += semana

with open('saida.html', 'w+') as f:
  f.write(html)
  
```
